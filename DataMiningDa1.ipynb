```python
import pandas as pd
from scipy import stats

data = pd.read_csv('Carseats.csv')
```

```python
print(data.head())
```

```python
numeric_columns = data.select_dtypes(include='number')
```

```python
summary = numeric_columns.describe()
min_values = numeric_columns.min()
max_values = numeric_columns.max()
mean_values = numeric_columns.mean()
median_values = numeric_columns.median()
mode_values = numeric_columns.mode().iloc[0]
quantiles = numeric_columns.quantile([0.25, 0.5, 0.75])
std_dev = numeric_columns.std()
variance = numeric_columns.var()
```

```python
print("\nSummary:\n", summary)
print("\nMinimum Values:\n", min_values)
print("\nMaximum Values:\n", max_values)
print("\nMean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
print("\nQuantiles:\n", quantiles)
print("\nStandard Deviation:\n", std_dev)
print("\nVariance:\n", variance)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Carseats.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Advertising'], data['Sales'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Advertising vs. Sales')
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
avg_sales_by_shelve = data.groupby('ShelveLoc')['Sales'].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_sales_by_shelve.plot(kind='barh', color='orange')
plt.title('Average Sales by Shelve Location')
plt.xlabel('Average Sales')
plt.ylabel('Shelve Location')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Sales'], bins=20, color='green', alpha=0.7)
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
avg_sales_by_price = data.groupby('Price')['Sales'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_sales_by_price.index, avg_sales_by_price.values, marker='o', color='purple')
plt.title('Line Graph of Average Sales by Price')
plt.xlabel('Price')
plt.ylabel('Average Sales')
plt.grid()
plt.show()
```
